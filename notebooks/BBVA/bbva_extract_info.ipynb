```python
#Función para buscar la pagína en la que se encuentra una palabra o texto
def find_page(path,pdf,word):
    doc = fitz.open(path+pdf)
    num_pages = doc.pageCount
    for i in range(num_pages):
        page = doc.loadPage(i)
        text = page.getText("text").replace('\n',' ')
        if text.find(word)!=-1:
            page = str(i+1)
            break 
    return page
```

```python
#Funcion para obtener el activo de BBVA con dos salidas
def get_activo_bbva(path,pdf,word):
    page = find_page(path, pdf, word)
    tabla_leer = camelot.read_pdf(path+pdf,flavor="stream",pages=page,edge_tol=30,row_tol=10)
    activo = tabla_leer[0].df.iloc[1:]
    activo_v = activo
    activo = activo[activo.columns[::-1]]
    activo.columns = ['T_0_feature_value',
                        'T_1_feature_value',
                        'T_2_feature_value',
                        'T_3_feature_value',
                        'T_4_feature_value',
                       'label']
    filename = str(pdf)
    period  = re.findall('[1-4]T[0-9]{2}',filename)[0]
    period = '20'+period[2:]+period[0:2][::-1]
    rows,cols = activo.shape
    #result_state
    tabla = [{'trimester_date': period,
     'bank_name': 'BBVA',
     'table_name': 'Balance General Activo',
     'short_descripction': activo.iloc[row,5],
     'T_0_feature_value': activo.iloc[row,0],
     'T_1_feature_value': activo.iloc[row,1],
     'T_2_feature_value': activo.iloc[row,2],
     'T_3_feature_value': activo.iloc[row,3],
     'T_4_feature_value': activo.iloc[row,4]} for row in range(rows)]
    return activo_v, pd.DataFrame.from_records(tabla)
    #return activo
```

```python
def get_pasivo_bbva(path, pdf, word):
    page = find_page(path, pdf, word)
    tabla_leer = camelot.read_pdf(path+pdf,flavor='stream',pages=page,flag_size=True,
                              columns=['0','5','20'])
    pasivo = tabla_leer[0].df#.str.split('\n',expand=True)
    pasivo = pasivo[1].str.split('\n', expand = True)
    pasivo = pasivo.iloc[1:]
    pasivo_v = pasivo 
    pasivo = pasivo[pasivo.columns[::-1]]
    pasivo.columns = ['T_0_feature_value',
                        'T_1_feature_value',
                        'T_2_feature_value',
                        'T_3_feature_value',
                        'T_4_feature_value',
                       'label']
    filename = str(pdf)
    period  = re.findall('[1-4]T[0-9]{2}',filename)[0]
    period = '20'+period[2:]+period[0:2][::-1]
    rows,cols = pasivo.shape
    tabla = [{'trimester_date': period,
     'bank_name': 'BBVA',
     'table_name': 'Balance General Activo',
     'short_descripction': pasivo.iloc[row,5],
     'T_0_feature_value': pasivo.iloc[row,0],
     'T_1_feature_value': pasivo.iloc[row,1],
     'T_2_feature_value': pasivo.iloc[row,2],
     'T_3_feature_value': pasivo.iloc[row,3],
     'T_4_feature_value': pasivo.iloc[row,4]} for row in range(rows)]
    return pasivo_v, pd.DataFrame.from_records(tabla) 
    #return pasivo
```

```python
def get_text_page_pypdf(path,pdf):
    pdfFileObject = open(path+pdf, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObject)
    count = pdfReader.numPages
    print(count)
    list_pages =[]
    for i in range(count):
        page = pdfReader.getPage(i)
        text = page.extractText()
        list_pages.append(text)
    text_pdf = " ".join(list_pages)
    #print(output)
    return {'Nombre_Archivo':pdf,
     'Texto':text_pdf,
    'Tipo':'Nativo'}
```
